```python
from transformers import TapasTokenizer, TapasForQuestionAnswering
import pandas as pd

# Initialiser le tokenizer et le modèle
tokenizer = TapasTokenizer.from_pretrained("google/tapas-base-finetuned-wtq")
model = TapasForQuestionAnswering.from_pretrained("google/tapas-base-finetuned-wtq")

# Créer un DataFrame avec les données
data = {
    "Product": ["Apples", "Bananas", "Oranges"],
    "Sales 2020": [120, 150, 100],
    "Sales 2021": [130, 160, 110],
}
df = pd.DataFrame(data).astype(str)  # Convertir toutes les données en chaînes de caractères

# Formuler une question en anglais
questions = ["What are the total sales of Bananas in 2020 and 2021?"]

# Préparer les inputs pour le modèle
inputs = tokenizer(table=df, queries=questions, padding="max_length", return_tensors="pt", truncation=True)

# Générer une réponse avec le modèle
outputs = model(**inputs)

# Obtenir les prédictions
predictions = tokenizer.convert_logits_to_predictions(
    inputs,
    outputs.logits.detach(),
    outputs.logits_aggregation.detach()
)

# Extraire les coordonnées de réponse et les indices d'agrégation
answer_coordinates = predictions[0]
answer_aggregation_indices = predictions[1] if len(predictions) > 1 else None

print("Answer Coordinates:", answer_coordinates)
# Pour afficher les indices d'agrégation, si disponibles
if answer_aggregation_indices is not None:
    print("Answer Aggregation Indices:", answer_aggregation_indices)
```
